### Budget to scale (1-4)

```python
has_empty_values = df['Budget'].isnull().any()

if has_empty_values:
    print("The column contains empty values.")
else:
    print("The column does not contain empty values.")
```

```python
has_empty_values = (df['Budget'] == '').any()
if has_empty_values:
    print("The column contains empty values.")
else:
    print("The column does not contain empty values.")
```

```python
df['Lunch price rate'] = 0
df['Dinner price rate'] = 0

# Custom function to encode the price categories
def encode_price_category(price):
    if price < 1000:
        return 1
    elif price < 3000:
        return 2
    elif price < 5000:
        return 3
    else:
        return 4

# Iterate over each row and update the DataFrame
for index, row in df.iterrows():
    lunch_prices = []
    dinner_prices = []

    if 'Lunch:' in row['Budget']:
        lunch_text = row['Budget'].split('Lunch:')[1].split('Dinner:')[0].strip()
        lunch_prices = [int(price.replace('￥', '').replace(',', '').split('～')[0]) for price in lunch_text.split() if '￥' in price and price.replace('￥', '').replace(',', '').split('～')[0].isdigit()]

    if 'Dinner:' in row['Budget']:
        dinner_text = row['Budget'].split('Dinner:')[1].strip()
        dinner_prices = [int(price.replace('￥', '').replace(',', '').split('～')[0]) for price in dinner_text.split() if '￥' in price and price.replace('￥', '').replace(',', '').split('～')[0].isdigit()]

    if lunch_prices:
        df.at[index, 'Lunch price rate'] = int(max(encode_price_category(price) for price in lunch_prices))

    if dinner_prices:
        df.at[index, 'Dinner price rate'] = int(max(encode_price_category(price) for price in dinner_prices))
```

### Transportation to meters from nearest station

```python
import re

def extract_meters(string):
    match = re.search(r'(\d+)\s*meters', string)
    if match:
        return int(match.group(1))
    else:
        return None

df['meters from nearest station'] = df['Transportation'].apply(extract_meters)
```
