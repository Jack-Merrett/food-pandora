```python
latitudes = []
longitudes = []
altitudes = []

# Iterate over each address in the column
for address in df['Addresses']:
    # Perform geocoding
    location = geolocator.geocode(address)
    
    # If a location is found, append the latitude, longitude, and altitude to the respective lists
    if location:
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
        altitudes.append(location.altitude)
    else:
        # If no location is found, append None values
        latitudes.append(None)
        longitudes.append(None)
        altitudes.append(None)

# Add the latitude, longitude, and altitude lists as new columns in the DataFrame
# df'latitude'] = latitudes
# df['longitude'] = longitudes
# df['altitude'] = altitudes
```
